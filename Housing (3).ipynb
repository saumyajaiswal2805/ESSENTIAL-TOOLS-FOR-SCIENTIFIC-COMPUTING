```python
# Dataset
import pandas as pd
import numpy as np
df = pd.read_csv(r'E:\MLBS\Assignment3\Tutorial_04-09-23\housing.csv')
df
#print(df.head())
```

```python
# Drop rows with missing values (if any)
df.dropna(inplace=True)
```

```python
X_data = df.drop(columns=['MEDV'])
y_data = df['MEDV']
X_data = X_data.to_numpy()
y_data = y_data.to_numpy()
X_data = (X_data-np.min(X_data))/(np.max(X_data)-np.min(X_data))
y_data = (y_data-np.min(y_data))/(np.max(y_data)-np.min(y_data))
print(X_data.shape)
print(y_data.shape)
```

```python
class NN:

    def __init__(self, input_layer_size=13, hidden_layer_size=2, output_layer_size=1, lr=0.01, epochs=1000):
        self.input_layer_size = input_layer_size
        self.hidden_layer_size = hidden_layer_size
        self.output_layer_size = output_layer_size
        self.lr = lr
        self.number_of_epochs = epochs

        # Initialize weights and biases randomly
        self.W1 = np.random.randn(self.input_layer_size, self.hidden_layer_size)
        self.b1 = np.random.randn(1, self.hidden_layer_size)
        self.W2 = np.random.randn(self.hidden_layer_size, self.output_layer_size)
        self.b2 = np.random.randn(1, self.output_layer_size)

    def activation(self, x):
        return 1 / (1 + np.exp(-x))

    def activation_derivative(self, x):
        return x * (1 - x)

    def forward(self, X):
        self.hidden = self.activation(np.dot(X, self.W1) + self.b1)
        self.output = self.activation(np.dot(self.hidden, self.W2) + self.b2)
        return self.output

    def back_prop(self, X, y):
        # Calculate error
        error = y - self.output

        # Compute gradients
        d_output = error * self.activation_derivative(self.output)
        d_hidden = np.dot(d_output, self.W2.T) * self.activation_derivative(self.hidden)

        # Update weights and biases
        self.W2 += self.lr * np.dot(self.hidden.T, d_output)
        self.b2 += self.lr * np.sum(d_output, axis=0, keepdims=True)
        self.W1 += self.lr * np.dot(X.T, d_hidden)
        self.b1 += self.lr * np.sum(d_hidden, axis=0, keepdims=True)

    def train(self, X, y):
        for _ in range(self.number_of_epochs):
            self.forward(X)
            self.back_prop(X, y)
            

# # Create an instance of the class
# model = NN(input_layer_size=X_data.shape[1], hidden_layer_size=num_hidden_neurons, output_layer_size=1, lr=learning_rate, epochs=1000)

# # Train the model
# model.train(X_data, y_data.reshape(-1, 1))
```

```python
 def train_and_evaluate(X_data, y_data, num_hidden_neurons, learning_rate):
    # Create an instance of the class
    model = NN(input_layer_size=X_data.shape[1], hidden_layer_size=num_hidden_neurons, 
               output_layer_size=1, lr=learning_rate, epochs=1000)
    
    # Train the model
    model.train(X_data, y_data.reshape(-1, 1))

    
    # Evaluate the model
    predictions = model.forward(X_data)
    mse = np.mean((predictions - y_data.reshape(-1, 1)) ** 2)

    return mse

# Cases to evaluate
cases = [
    {"hidden_neurons": 3, "learning_rate": 0.01},
    {"hidden_neurons": 4, "learning_rate": 0.001},
    {"hidden_neurons": 5, "learning_rate": 0.0001}
]

# Train and evaluate each case
for i, case in enumerate(cases):
    mse = train_and_evaluate(X_data, y_data, case["hidden_neurons"], case["learning_rate"])
    print(f"Case {i+1}: MSE = {mse}")
```

```python
def k_fold_cross_validation(X, y, num_hidden_neurons, learning_rate, k=5):
    fold_size = len(X) // k
    mse_values = []

    for i in range(k):
        start_idx = i * fold_size
        end_idx = start_idx + fold_size

        X_train = np.concatenate((X[:start_idx], X[end_idx:]), axis=0)
        y_train = np.concatenate((y[:start_idx], y[end_idx:]), axis=0)

        X_val = X[start_idx:end_idx]
        y_val = y[start_idx:end_idx]

        model = NN(input_layer_size=X.shape[1], hidden_layer_size=num_hidden_neurons, output_layer_size=1, lr=learning_rate, epochs=1000)
        model.train(X_train, y_train)
        mse = np.mean((model.forward(X_val) - y_val) ** 2)
        mse_values.append(mse)

    return np.mean(mse_values)

# Perform five-fold cross-validation
for i, case in enumerate(cases):
    mse = k_fold_cross_validation(X_data, y_data.reshape(-1, 1), case["hidden_neurons"], case["learning_rate"], k=5)
    print(f"Five-fold CV - Case {i+1}: Mean MSE = {mse}")
```
